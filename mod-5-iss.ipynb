```python
import time
import matplotlib.pyplot as plt
import requests

xvalues=[]
yvalues=[]

for _ in range(100):
    data = requests.get("http://api.open-notify.org/iss-now.json").json()
    xvalues.append(data["iss_position"]["latitude"])
    yvalues.append(data["iss_position"]["longitude"])
    
    plt.title("Scatterplot of the ISS at time " + str(data["timestamp"]))
    plt.scatter(xvalues,yvalues,label="ISS", color="g")
    plt.xlabel('Latitude')
    plt.ylabel('Longitude')
    leg = plt.legend(loc='upper left')
    plt.show()            
                
    time.sleep(15)
```
